# joining NumPy Array

```python
import numpy as np
a=np.array([10,20,30])
b=np.array([40,80,90])
np.vstack((a,b))
```

```python
np.hstack((a,b))
```

```python
np.hstack((a,b))
```

```python
np.column_stack((a,b))
```

# NumPy intersection and differences 

```python
x=np.array([10,20,30,40,50,60])
y=np.array([30,50,90,100,120,150])
np.intersect1d(x,y)
```

```python
np.setdiff1d(x,y)
```

# Addition of NumPy Arrays

```python
t=np.array([10,20])
s=np.array([30,40])
np.sum([t,s])
```

```python
np.sum([t,s],axis=0)  #adds column
```

```python
np.sum([t,s],axis=1)  #adds rows
```
